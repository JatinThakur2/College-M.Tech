```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load Iris dataset
iris = sns.load_dataset("iris")

# Select numeric columns only
numeric_cols = iris.select_dtypes(include="number").columns

# Loop through numeric columns
for col in numeric_cols:
    data = iris[col]

    # --------------------------
    # Calculate Central Tendency
    # --------------------------
    mean = data.mean()
    median = data.median()
    mode = data.mode()[0]  # pick first mode if multiple

    # --------------------------
    # Calculate Dispersion
    # --------------------------
    data_range = data.max() - data.min()
    std_dev = data.std()
    range_min = data.min()
    range_max = data.max()

    # --------------------------
    # 1. Histogram with Mean, Median, Mode
    # --------------------------
    plt.figure(figsize=(8,5))
    sns.histplot(data, bins=15, kde=True, color="skyblue", stat="density")
    plt.axvline(mean, color='r', linestyle='--', label=f'Mean={mean:.2f}')
    plt.axvline(median, color='g', linestyle='--', label=f'Median={median:.2f}')
    plt.axvline(mode, color='b', linestyle='--', label=f'Mode={mode:.2f}')
    plt.title(f"{col.capitalize()} Distribution - Mean, Median, Mode")
    plt.xlabel(col)
    plt.ylabel("Density")
    plt.legend()
    plt.show()

    # --------------------------
    # 2. Line Plot Showing Range, Mean, Median, 1 Std Dev
    # --------------------------
    fig, ax = plt.subplots(figsize=(10,5))
    
    # Plot data points
    ax.plot(data, marker='o', linestyle='-', color='skyblue', label='Data Points')
    
    # Horizontal lines for key stats
    ax.axhline(y=range_min, color='r', linestyle='--', label=f'Min={range_min:.2f}')
    ax.axhline(y=range_max, color='r', linestyle='--', label=f'Max={range_max:.2f}')
    ax.axhline(y=mean, color='g', linestyle='--', label=f'Mean={mean:.2f}')
    ax.axhline(y=median, color='b', linestyle='--', label=f'Median={median:.2f}')
    ax.axhline(y=mean + std_dev, color='purple', linestyle=':', label=f'Mean + 1 Std Dev')
    ax.axhline(y=mean - std_dev, color='purple', linestyle=':', label=f'Mean - 1 Std Dev')
    
    ax.set_title(f"{col.capitalize()} - Range, Mean, Median, Std Dev")
    ax.set_xlabel("Data Point Index")
    ax.set_ylabel(f"{col} Value")
    ax.legend()
    plt.show()

    # --------------------------
    # Print statistics in console
    # --------------------------
    print(f"\n===== {col.upper()} =====")
    print(f"Minimum Value: {range_min:.2f}")
    print(f"Maximum Value: {range_max:.2f}")
    print(f"Range: {data_range:.2f}")
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Mode: {mode:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")
```
