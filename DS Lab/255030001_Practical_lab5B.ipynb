```python
# Implement a simple linear regression using the following Salary dataset:
# https://github.com/content-anu/dataset-simple-linear

# Perform the following:
# Step 1: Importing the dataset
# Step 2: Data pre-processing
# Step 3: Splitting the test and train sets
# Step 4: Fitting the linear regression model to the training set
# Step 5: Predicting test results
# Step 6: Visualizing the test results

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
# Step 1: Importing the dataset
url = "https://raw.githubusercontent.com/content-anu/dataset-simple-linear/master/Salary_Data.csv"
dataset = pd.read_csv(url)      
X = dataset.iloc[:, :-1].values  # Independent variable (Years of Experience)
y = dataset.iloc[:, 1].values    # Dependent variable (Salary)  
print("table", dataset.head())
```

```python
# Step 2: Data pre-processing
# Check for missing values
if dataset.isnull().sum().any():
    dataset = dataset.dropna()  # Drop missing values if any
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values  
print("Data pre-processing completed. Dataset shape:", dataset.shape)
```

```python
# Step 3: Splitting the test and train sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print("Training set size:", X_train.shape[0])
print("Test set size:", X_test.shape[0])
```

```python
# Step 4: Fitting the linear regression model to the training set
regressor = LinearRegression()
regressor.fit(X_train, y_train)
print("Model coefficients:", regressor.coef_)
print("Model intercept:", regressor.intercept_)
print("slope:", regressor.coef_[0])
print("Model trained successfully.")
```

```python
# Step 5: Predicting test results
y_pred = regressor.predict(X_test)
dataframe = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print("Predicted salaries:", y_pred, y_test)
```

```python
# Step 6: Visualizing the test results
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, y_pred, color='blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
##check model r squared and mean squared error
from sklearn.metrics import r2_score, mean_squared_error
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print("R-squared:", r2)
print("Mean Squared Error:", mse)
```
