```python
## Create ten different versions of line charts for a particular set of (x, y) points.
## Which ones are best and which ones worst? Explain why
#Ans: Among the ten line-chart variations, the most effective are the clean and honest designs—specifically charts 1 through 5 (and chart 6 if a dark background is desired). These versions keep a clear axis starting at zero, use simple colors, and maintain a high data-ink ratio so the viewer’s attention stays on the actual trend. Charts 7 and 8 fall in the middle: the heavy grid in chart 7 competes with the data for attention, while the duplicate series in chart 8 adds complexity without adding information. The poorest choices are charts 9 and 10. Chart 9 begins the y-axis above zero, exaggerating small differences and giving a misleading sense of change. Chart 10 clutters the view with bright colors, decorative backgrounds, and oversized fonts, distracting from the data itself. Overall, the best charts communicate clearly and truthfully, while the worst either distort the data or bury it beneath unnecessary decoration.

import numpy as np
import matplotlib.pyplot as plt

# Simple x–y dataset
x = np.array([0, 1, 2, 3, 4])
y = np.array([1, 3, 2, 5, 4])


```

```python
# 1. Clean default
plt.figure()
plt.plot(x, y)
plt.title("1. Clean Default")
plt.show()
```

```python
# 2. With markers
plt.figure()
plt.plot(x, y, marker='o')
plt.title("2. Markers")
plt.show()
```

```python
# 3. Dashed line
plt.figure()
plt.plot(x, y, linestyle='--', color='purple')
plt.title("3. Dashed Line")
plt.show()
```

```python
# 5. Shaded area under curve
plt.figure()
plt.fill_between(x, y, color='skyblue', alpha=0.4)
plt.plot(x, y, color='blue')
plt.title("5. Area Under Curve")
plt.show()
```

```python
# 6. Dark theme
plt.style.use('dark_background')
plt.figure()
plt.plot(x, y, marker='D', color='orange')
plt.title("6. Dark Theme")
plt.show()
plt.style.use('default')
```

```python
# 7. Heavy grid
plt.figure()
plt.plot(x, y, marker='s')
plt.grid(True, linewidth=1.5, color='black')
plt.title("7. Heavy Grid")
plt.show()
```

```python
# 8. Multiple series (same data twice, different style)
plt.figure()
plt.plot(x, y, marker='o', label="Line A")
plt.plot(x, y + 0.5, linestyle='--', label="Line B")
plt.legend()
plt.title("8. Multiple Series")
plt.show()
```

```python
# 9. Step y-axis
plt.figure()
plt.step(x, y, marker='o')
plt.title("9. Distorted Axis (Misleading)")
plt.show()
```

```python
# 10. Cluttered decoration
plt.figure(facecolor='pink')
plt.plot(x, y,
         marker='D',
         linewidth=4,
         linestyle='--',
         color='purple')

# Add value labels at each diamond
for xi, yi in zip(x, y):
    plt.text(xi, yi + 0.3,
             f"({xi}, {yi})",
             ha='center',
             fontsize=10,
             color='black',
             weight='bold')

plt.grid(True, linestyle=':', linewidth=2, color='green')
plt.xlabel("X Values", fontsize=14, color='red')
plt.ylabel("Y Values", fontsize=14, color='red')
plt.title("10. Clip-Art Clutter", fontsize=18, color='blue', weight='bold')
plt.show()



```
