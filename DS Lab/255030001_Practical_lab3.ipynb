```python
##---------Question 1. Read a text file, and count the total number of words. ----------------

def count(filepath):
    with open(filepath, 'r') as file:
        content = file.read()
        words = content.split()
        return len(words)
file_path = 'simple.txt'
word_count = count(file_path)
print(word_count)
```

```python
##---------Question 2. Read a text file, and count the total number of distinct words.--------

def count(filepath):
    with open(filepath, 'r') as file:
        content = file.read()
        words = content.split()
        distinct =set(words)
        return len(distinct)
file_path = 'simple.txt'
word_count = count(file_path)
print(word_count)
```

```python
##--------Question 3. Read a file of numbers, and plot a frequency histogram of them.----------------
import pandas as pd
import matplotlib.pyplot as plt

def plot_histogram(filepath):
   
    df = pd.read_csv(filepath, delimiter=',')
    values = df['Value']
    plt.hist(values, bins=5, edgecolor='black')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Frequency Histogram of Values')
    plt.show()

file_path = 'data.txt'  
plot_histogram(file_path)

```
