```python
##-------------Practice from geek--------------------------------

##-------------Scatter Plot--------------------------------------
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# reading the database
data = pd.read_csv("tips.csv")
display(data.head(10))

# ------------------- Scatter Plot -------------------
plt.figure(figsize=(6,4))
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=data)
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()

# ------------------- Heatmap -------------------
plt.figure(figsize=(6,4))
numeric_data = data.select_dtypes(include=['float64', 'int64'])  # only numbers
sns.heatmap(numeric_data.corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap of Correlation (Numeric Columns Only)")
plt.show()


# ------------------- Bar Plot -------------------
plt.figure(figsize=(6,4))
sns.barplot(x='day', y='total_bill', data=data, estimator=sum, errorbar=None)
plt.title("Bar Plot: Total Bill by Day")
plt.show()


# ------------------- Histogram -------------------
plt.figure(figsize=(6,4))
sns.histplot(x='total_bill', data=data, kde=True, hue='sex')
plt.title("Histogram of Total Bill")
plt.show()

# ------------------- Data Map (Simulated) -------------------
np.random.seed(0)
data['lat'] = np.random.uniform(37, 42, size=len(data))   # Somewhere in USA
data['lon'] = np.random.uniform(-122, -117, size=len(data))

plt.figure(figsize=(6,6))
plt.scatter(data['lon'], data['lat'], c=data['total_bill'], cmap='viridis', s=50, alpha=0.7)
plt.colorbar(label="Total Bill")
plt.title("Simulated Data Map of Bills")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

# ------------------- Pie Chart -------------------
plt.figure(figsize=(6,6))
bill_by_day = data.groupby('day')['total_bill'].sum()

plt.pie(bill_by_day, labels=bill_by_day.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title("Pie Chart: Distribution of Total Bill by Day")
plt.show()

```
