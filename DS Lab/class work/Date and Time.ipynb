```python
import datetime

now = datetime.datetime.now()
print("Current Date & Time:", now)
print("================================")

# --- Year ---
print("%Y (Year, full):", now.strftime("%Y"))   # 2025
print("%y (Year, short):", now.strftime("%y"))  # 25

# --- Month ---
print("%m (Month 01-12):", now.strftime("%m"))  # 09
print("%B (Month full name):", now.strftime("%B")) # September
print("%b (Month short name):", now.strftime("%b")) # Sep

# --- Day ---
print("%d (Day of month 01-31):", now.strftime("%d")) 
print("%j (Day of year 001-366):", now.strftime("%j")) 

# --- Week ---
print("%U (Week number, Sunday first):", now.strftime("%U")) 
print("%W (Week number, Monday first):", now.strftime("%W")) 
print("%w (Weekday as number, Sunday=0):", now.strftime("%w")) 
print("%A (Weekday full name):", now.strftime("%A")) 
print("%a (Weekday short name):", now.strftime("%a")) 

# --- Time ---
print("%H (Hour 00-23):", now.strftime("%H"))
print("%I (Hour 01-12):", now.strftime("%I"))
print("%M (Minute 00-59):", now.strftime("%M"))
print("%S (Second 00-59):", now.strftime("%S"))
print("%f (Microsecond 000000-999999):", now.strftime("%f"))
print("%p (AM/PM):", now.strftime("%p"))
print("%z (UTC offset):", now.strftime("%z"))
print("%Z (Timezone name):", now.strftime("%Z"))

# --- Combined Formats ---
print("%c (Locale date & time):", now.strftime("%c")) 
print("%x (Locale date):", now.strftime("%x")) 
print("%X (Locale time):", now.strftime("%X")) 
print("%r (12-hour clock with AM/PM):", now.strftime("%r")) 
```

```python
import datetime

# Capture the First Date
day1 = datetime.date(2018, 2, 12)
print("day1:", day1.ctime())

# Capture the Second Date
day2 = datetime.date(2017, 8, 18)
print("day2:", day2.ctime())

# Find the difference between the dates
print("Number of Days:", day1 - day2)

# Today's date
date_today = datetime.date.today()

# Create a delta of Four Days
no_of_days = datetime.timedelta(days=4)

# Use Delta for Past Date
before_four_days = date_today - no_of_days
print("Before Four Days:", before_four_days)

# Use Delta for Future Date
after_four_days = date_today + no_of_days
print("After Four Days:", after_four_days)
```

```python
import datetime

date_today = datetime.date.today()
print("Today is:", date_today)

# Create a delta of Four Days 
no_of_days = datetime.timedelta(days=4) 

# Use Delta for Past Date
before_four_days = date_today - no_of_days 
print("Before Four Days:", before_four_days) 

# Use Delta for Future Date
after_four_days = date_today + no_of_days

date1 = datetime.date(2018, 4, 4)
print("date1:", date1)

# Comparisons
if date1 == before_four_days:
    print("Same Dates")
if date_today > date1:
    print("Past Date")
if date1 < after_four_days:
    print("Future Date")
```

```python
import pandas as pd

# Left DataFrame
left = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
    'subject_id': ['sub1', 'sub2', 'sub4', 'sub6', 'sub5']
})

# Right DataFrame
right = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
    'subject_id': ['sub2', 'sub4', 'sub3', 'sub6', 'sub5']
})

print("Left DataFrame:\n", left)
print("\nRight DataFrame:\n", right)

# Merge on "id"
merged = pd.merge(left, right, how="inner", on="id")

print("\nMerged DataFrame (on 'id'):\n", merged)
```

```python
import pandas as pd

one = pd.DataFrame({
    'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
    'subject_id': ['sub1', 'sub2', 'sub4', 'sub6', 'sub5'],
    'Marks_scored': [98, 90, 87, 69, 78]
}, index=[1, 2, 3, 4, 5])

two = pd.DataFrame({
    'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
    'subject_id': ['sub2', 'sub4', 'sub3', 'sub6', 'sub5'],
    'Marks_scored': [89, 80, 79, 97, 88]
}, index=[1, 2, 3, 4, 5])

# Concatenate DataFrames
result = pd.concat([one, two])

print("Concatenated DataFrames:\n", result)
```
