```python
# 1. Write a Python program to create a vector of a specified type and length. Create a vector of numeric, complex, logical, and character types of length 6.

num_v = [float(i) for i in range(6)]
comp_v = [complex(i, i+1) for i in range(6)]
log_v = [i % 2 == 0 for i in range(6)]
char_v = [chr(65 + i) for i in range(6)]  # ['A', 'B', 'C', 'D', 'E', 'F']

print("Numeric Vector:", num_v)
print("Complex Vector:", comp_v)
print("Logical Vector:", log_v)
print("Character Vector:", char_v)
```

```python
#2. Add two vectors of integer type and length 3
vector1 = [1,2,4]
vector2 = [3,4,6]
sum_vector = [vector1[i] + vector2[i] for i in range(3)]

print("Vector 1:", vector1)
print("Vector 2:", vector2)
print("Sum Vector:", sum_vector)
```

```python
# 3. Create a list with a vector, matrix, and list, then remove the second element
import numpy as np

vector = [1, 2, 3]
matrix = np.array([[1, 2], [3, 4]])
list = ['a', 'b', 'c']
combined_list = [vector, matrix, list]

print("combined List:", combined_list)
del combined_list[1]

print("Modified List (after removing second element):", combined_list)

```

```python
# 4. Create a list with a vector, matrix, and list, then update the last element
import numpy as np
vector = [1, 2, 3]
matrix = np.array([[1, 2], [3, 4]])
list = ['a', 'b', 'c']
combined_list = [vector, matrix, list]
print("Combined List: ",combined_list)

combined_list[-1] = ['x', 'y', 'z']

print("Modified List (after updating last element):", combined_list)

```
